```python
import csv
import numpy as np
from sklearn.datasets import load_iris
from sklearn.neighbors import KNeighborsClassifier

data = []
with open("C:\data\irisDataset.csv","r") as csvfile:
    reader = csv.reader(csvfile,delimiter = ',')
    for row in reader:
        data.append(row)

lengthData = len(data)


inputData = np.zeros((lengthData,4))


for i in range(len(data)):
    for j in range(4):
        inputData[i][j] = data[i][j]

outputData = np.zeros(lengthData)
for i in range(len(data)):
    outputData[i] = data[i][4]

print("Input Data is")
print(inputData)

print("")
print("Output Data is")
print(outputData)
        
print("Data is ")
print(data)

iris = load_iris()
print(type(iris))
print(iris.data)
print(iris.feature_names)
print(iris.target)
print(iris.target_names)

x = iris.data
y = iris.target

knn = KNeighborsClassifier(n_neighbors=1)
print(knn)
knn.fit(x,y)
predictInput = np.array([3,5,4,2])
predictInput = predictInput.reshape(1,-1)
print("The estimated type is ")
print(knn.predict(predictInput))

X_new = np.array([[3,5,4,2],[5,4,3,2]])
print(knn.predict(X_new))
```

```python
print("FROM VIDEO LECTURE 4")

print("**********************************************************************************************")
print("**********************************************************************************************")
print("**********************************************************************************************")
print("**********************************************************************************************")
print("**********************************************************************************************")

import csv
import numpy as np
import scipy as sp
from sklearn.datasets import load_iris
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression

data = []
with open("C:\data\irisDataset.csv","r") as csvfile:
    reader = csv.reader(csvfile,delimiter = ',')
    for row in reader:
        data.append(row)

lengthData = len(data)

inputData = np.zeros((lengthData,4))

for i in range(len(data)):
    for j in range(4):
        inputData[i][j] = data[i][j]

outputData = np.zeros(lengthData)
for i in range(len(data)):
    outputData[i] = data[i][4]

print("Input Data is")
print(inputData)

print("")
print("Output Data is")
print(outputData)
        
print("Data is ")
print(data)

iris = load_iris()
print(type(iris))
print(iris.data)
print(iris.feature_names)
print(iris.target)
print(iris.target_names)

x = iris.data
y = iris.target

knn = KNeighborsClassifier(n_neighbors=1)
print(knn)
knn.fit(x,y)
predictInput = np.array([3,5,4,2])
predictInput = predictInput.reshape(1,-1)
print("The estimated type is ")
print(knn.predict(predictInput))

X_new = np.array([[3,5,4,2],[5,4,3,2]])
print(knn.predict(X_new))

logreg = LogisticRegression()
logreg.fit(x,y)
print("logistic regression predictions :")
print(logreg.predict(predictInput))
print(logreg.predict(X_new))
```
