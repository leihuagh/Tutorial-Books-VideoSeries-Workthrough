```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

from sklearn.datasets import load_iris
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn.cross_validation import train_test_split

#Path for file Advertising.csv
#data = pd.read_csv("http://www-bcf.usc.edu/~gareth/ISL/Advertising.csv")
#if proxy problem occurs, same file is available in repo.Give the appropriate address 
data = pd.read_csv("C:\\Users\\sj141366\\jupyter notebook\\Scikit_Learn Kaggle Video Tutorial\\Advertising.csv",index_col = 0)

print(data.head(5))
print(data.tail())
print(data.shape)

#Plotting input features individually with output 
#sns.pairplot(data,x_vars=['TV','Radio','Newspaper'],y_vars='Sales',size=5,aspect=0.7)

#Regression is done to fit data into best possible value
#sns.pairplot(data,x_vars=['TV','Radio','Newspaper'],y_vars='Sales',size=5,aspect=0.7,kind='reg')

#List of features 
feature_cols = ['TV','Radio','Newspaper']

#Create Dataframe of Input Data
X = data[feature_cols]

#or one can combine these all
#X = data[['TV','Radio','Newspaper']]

print(X.head())
print(type(X))
print(X.shape)

#Same as above for output data
y = data['Sales']

#some other way - only works if no spaces between column name
y = data.Sales

print(y.head())
print(type(y))
print(y.shape)

#Since no size for testig set is mentioned, default size is 25%
XTrain,XTest,yTrain,yTest = train_test_split(X,y,random_state=1)

print(XTrain.shape)
print(XTest.shape)
print(yTrain.shape)
print(yTest.shape)

linreg = LinearRegression()

linreg.fit(XTrain,yTrain)

print(linreg.coef_)
print(linreg.intercept_)

#pairing feature names with feature coefficient
print(zip(feature_cols,linreg.coef_))
#y = 2.87696662232 + 0.04656457 * TV + 0.17915812 * Radio + 0.00345046 * Newspaper

#Prediction for Test set data
yPred = linreg.predict(XTest)


#Model Evaluation metrices for regression
#let true and predicted response is
true = [100,50,30,20]
pred = [90,50,50,30]

#1.Mean Absolute error => (1/n) * summation(|ytrue - ypredict|)
#=> (10+0+20+10)/4 = 10
#now with scikitLearn
MAEAccuracy = metrics.mean_absolute_error(true,pred)
print(MAEAccuracy) #Ans = 10.0

#2.Mean squared error => ((1/n) * summation((|(ytrue) - (ypredict)|)^2))
#=> ((100+0+400+100)/4) = (150)
#now with scikitLearn
MSEAccuracy = metrics.mean_squared_error(true,pred)
print(MSEAccuracy) #Ans = 150

#3.Root Mean squared error => sqrt((1/n) * summation((|(ytrue) - (ypredict)|)^2))
#=> sqrt((100+0+400+100)/4) = sqrt(150)
#now with scikitLearn
RMSEAccuracy = np.sqrt(metrics.mean_squared_error(true,pred))
print(RMSEAccuracy) #Ans = 12.2474487139


#Going to actual Example
modelAccuracy = np.sqrt(metrics.mean_squared_error(yTest,yPred))
print(modelAccuracy)#Ans = 1.40465142303

#Repeating steps by removing Newspaper column from Data to check how it affects 
#the Accuracy
feature_cols = ['TV','Radio']
X = data[feature_cols]
XTrain,XTest,yTrain,yTest = train_test_split(X,y,random_state=1)
y = data.Sales
linreg.fit(XTrain,yTrain)
yPred = linreg.predict(XTest)
modelAccuracy = np.sqrt(metrics.mean_squared_error(yTest,yPred))
print(modelAccuracy)#Ans = 1.38790346994

#Repeating steps by removing Radio column from Data to check how it affects 
#the Accuracy
feature_cols = ['TV','Newspaper']
X = data[feature_cols]
XTrain,XTest,yTrain,yTest = train_test_split(X,y,random_state=1)
y = data.Sales
linreg.fit(XTrain,yTrain)
yPred = linreg.predict(XTest)
modelAccuracy = np.sqrt(metrics.mean_squared_error(yTest,yPred))
print(modelAccuracy)#Ans = 3.35205633443

#Repeating steps by removing TV column from Data to check how it affects 
#the Accuracy
feature_cols = ['Radio','Newspaper']
X = data[feature_cols]
XTrain,XTest,yTrain,yTest = train_test_split(X,y,random_state=1)
y = data.Sales
linreg.fit(XTrain,yTrain)
yPred = linreg.predict(XTest)
modelAccuracy = np.sqrt(metrics.mean_squared_error(yTest,yPred))
print(modelAccuracy)#Ans = 3.74869749747
```
